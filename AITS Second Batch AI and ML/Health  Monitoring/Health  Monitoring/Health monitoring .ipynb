```python
#to checking accuracy
from sklearn.metrics import accuracy_score
```

```python
import sklearn.metrics as metrics
```

```python
fpr,tpr,threshlod=metrics.roc_curve(y_test,y_predict)
```

```python
#roc while give all attritues like fpr...
#we use tpr,fpr b/z to pretict and either in two ways
roc_auc=metrics.auc(fpr,tpr)
                   
```

```python
roc_auc
```

```python

plt.plot(fpr,tpr,'b',label='AUC=%0.2f'%roc_auc,color='green')
plt.plot([0,1],[0,1],'r--')
plt.xlim([0,1])
plt.ylim([0,1])
plt.ylabel("true positive rate")
plt.xlabel("false positive rate")
plt.title("Recevier operating characterstics")
plt.legend(loc='lower right')
plt.show()
```
